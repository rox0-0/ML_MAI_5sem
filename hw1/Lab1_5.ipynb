6. Рассмотрим случайную матрицу размером 10х2, представляющую декартовы координаты, преобразуем их в полярные координаты

````python
coords = np.random.random((10, 2)) * 10
x, y = coords[:, 0], coords[:, 1]

r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)
print("Декартовы координаты:")
print(coords)
print("\nПолярные координаты (радиус, угол):")
polar_coords = np.column_stack((r, theta))
print(polar_coords)
````

7. Создайте структурированный массив с координатами x и y, охватывающий область [0,1]x[0,1]

````python
n_points = 5
x_coords = np.linspace(0, 1, n_points)
y_coords = np.linspace(0, 1, n_points)

X, Y = np.meshgrid(x_coords, y_coords)
coord_dtype = np.dtype([('x', float), ('y', float)])
coords_array = np.empty((n_points, n_points), dtype=coord_dtype)
coords_array['x'] = X
coords_array['y'] = Y

print("Структурированный массив координат :")
print(coords_array)
````

````python
import pandas as pd
````

````python
# Load dataset here
df = pd.read_csv('data.csv')
print(df.head())
````

1. Узнайте сколько было на борту человек

````python
people_count = len(df)
print("На борту было", people_count, "человек")
````

2. Какой медианный и средний возраста пассажиров

````python
median_age = df['Age'].median()
mean_age = df['Age'].mean()
print(f"Медианный возраст: {median_age}")
print(f"Средний возраст: {mean_age}")
````

3. Посчитайте процент выживаемости детей(до 16 лет) и взрослых

````python
children = df[df['Age'] < 16]
adults = df[df['Age'] >= 16]

children_survival_rate = (children['Survived'].sum() / len(children)) * 100
adults_survival_rate = (adults['Survived'].sum() / len(adults)) * 100


print(f"Процент выживаемости детей: {children_survival_rate:.2f}%")
print(f"Процент выживаемости взрослых: {adults_survival_rate:.2f}%")
````

4. Верна ли гипотеза, что женщины и дети сажались в шлюпки первыми и выживали больше?

````python
df['Category'] = 'man'
df.loc[df['Age'] <= 16, 'Category'] = 'child'
df.loc[(df['Sex'] == 'female') & (df['Age'] > 16), 'Category'] = 'woman'

survival_by_category = df.groupby('Category')['Survived'].mean() * 100
print("Выживаемость по категориям:")
print(survival_by_category)

women_child_survival = survival_by_category[['woman', 'child']].mean()
men_survival = survival_by_category['man']
print(f"\nСредняя выживаемость женщин и детей: {women_child_survival:.2f}%")
print(f"Выживаемость мужчин: {men_survival:.2f}%")
print(f"Гипотеза подтверждается: {women_child_survival > men_survival}")
````

5. Зависит ли выживаемость от класса обслуживания?

````python

survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print("Выживаемость по классам:")
print(survival_by_class)


diff_1_3 = survival_by_class[1] - survival_by_class[3]
print(f"\nРазница первого и третьего класса: {diff_1_3:.1f}%")

if diff_1_3 > 30:
    print("Зависимость есть")
else:
    print("Зависимости нет или она слабая")
````

6. Посчитайте средний возраст умерших женщин и мужчин

````python

dead_ages = df[df['Survived'] == 0].groupby('Category')['Age'].mean()
print("Средний возраст умерших женщин",dead_ages['woman'])
print("Средний возраст умерших мужчин",dead_ages['man'])
````

7. Различается ли процент выживаемости пассажиров с хотя бы одним родственником на борту и одиночек?

````python

df['HasRelatives'] = (df['SibSp'] + df['Parch']) > 0

survival_by_relatives = df.groupby('HasRelatives')['Survived'].mean() * 100
print("Выживаемость:")
print(f"С родственниками: {survival_by_relatives[True]:.1f}%")
print(f"Одиночки: {survival_by_relatives[False]:.1f}%")


diff = survival_by_relatives[True] - survival_by_relatives[False]
print(f"\nРазница: {diff:.1f}%")

if abs(diff) > 10:
    print("Значимая разница")
else:
    print("Разница незначительная")
````

8. Различается ли средняя стоимость билета у умерших и выживших пассажиров?

````python
fare_by_survival = df.groupby('Survived')['Fare'].mean()
print("Средняя стоимость билета:")
print(f"Выжившие: {fare_by_survival[1]:.2f}")
print(f"Умершие: {fare_by_survival[0]:.2f}")
````

Ответ - различается


9. Выведите максимальный и минимальный возраст выживших пассажиров во 2 классе обслуживания

````python
survived_2nd_class = df[(df['Survived'] == 1) & (df['Pclass'] == 2)]
max_age = survived_2nd_class['Age'].max()
min_age = survived_2nd_class['Age'].min()
print(f"Максимальный возраст выживших во 2 классе: {max_age}")
print(f"Минимальный возраст выживших во 2 классе: {min_age}")
````

10. Подтвердите или опровергните гипотезу: молодым мужчинам(от 18 лет) выжить легче, чем более взрослым

````python
men = df[df['Sex'] == 'male']
young_men = men[(men['Age'] >= 18) & (men['Age'] <= 30)]
adult_men = men[men['Age'] > 30]

young_survival = young_men['Survived'].mean() * 100
adult_survival = adult_men['Survived'].mean() * 100

print(f"Выживаемость молодых мужчин: {young_survival:.2f}%")
print(f"Выживаемость взрослых мужчин: {adult_survival:.2f}%")
print(f"Гипотеза подтверждается: {young_survival > adult_survival}")
````

6. Сравните плотности распределения возростов выживших и умерших пассажиров, сделайте выводы

````python
import seaborn as sns
plt.figure(figsize=(14, 7))

survived = df[df['Survived'] == 1]['Age'].dropna()
died = df[df['Survived'] == 0]['Age'].dropna()

sns.kdeplot(survived, fill=False, label='Выжившие', alpha=0.5, color='green', linewidth=2.5, bw_adjust=0.8)
sns.kdeplot(died, fill=False, label='Умершие', alpha=0.5, color='red', linewidth=2.5, bw_adjust=0.8)

plt.xlabel('Возраст', fontsize=12)
plt.ylabel('Плотность', fontsize=12)
plt.title('Плотности вероятностей выживших и умерших',
          fontsize=14, fontweight='bold', pad=15)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()

````

Выводы:
 В молодом возрасте распределение выживыших имеет более яркий пик, который показывает приоритет на спасение детей, в то время как распределение умерших более плоское и равномерное. так же видно, что средний возраст умерших больше среднего возраста выживших.


7. Постройте круговую диаграмму отображающую выживаемость мужчин и женщин в разных классах обслуживания(визуализация должна быть интуитивно понятная без объяснений, женщины и мужчины соответствующих классов объеденены визуально в одну группу и 2 подгруппы)*

\* Задание на доп. балл

````python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, pclass in enumerate([1, 2, 3]):
    class_data = df[df['Pclass'] == pclass]
    survival_counts = class_data.groupby(['Sex', 'Survived']).size().unstack(fill_value=0)
    survived_by_sex = survival_counts[1] if 1 in survival_counts.columns else pd.Series([0, 0], index=['female', 'male'])
    labels = ['Женщины', 'Мужчины']

    wedges, texts, autotexts = axes[i].pie(survived_by_sex,
                                          labels=labels,
                                          autopct='%1.1f%%',
                                          colors=['pink', 'blue'],
                                          startangle=90,
                                          textprops={'fontsize': 11})
    axes[i].text(0, -1.3, f'Класс {pclass}',
                ha='center', va='center',
                fontsize=14, fontweight='bold')

plt.suptitle('Выживаемость мужчин и женщин в разных классах обслуживания',
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````
